```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
df = pd.read_csv('salary.csv')
df = df['salary_in_usd']
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df, bins=20, density=True, alpha=0.6, color="g", label="Salaries")

# Расчет среднего и стандартного отклонения
mu, std = df.mean(), df.std()

# Определение диапазона значений для построения кривой нормального распределения
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Построение кривой нормального распределения
plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

# Настройка графика
plt.title("Distribution of salaries")
plt.xlabel("Salaries")
plt.ylabel("Density")
plt.legend()

# Отображение графика
plt.show()
```

```python
# Расчет математического ожидания
mean_value = df.mean()
print(f"Математическое ожидание: {mean_value:.3f} USD")

# Расчет дисперсии
variance = df.var()
print(f"Дисперсия: {variance:.3f} USD²")

# Расчет асимметрии
skewness = df.skew()
print(f"Асимметрия: {skewness:.3f}")

# Расчет эксцесса
kurtosis = df.kurtosis()
print(f"Эксцесс: {kurtosis:.3f}")

# Расчет квантилей
quantile_05 = df.quantile(0.05)
quantile_95 = df.quantile(0.95)
print(f"Квантиль 0,05: {quantile_05:.3f} USD")
print(f"Квантиль 0,95: {quantile_95:.3f} USD")

quantile_025 = df.quantile(0.025)
print(f"2,5%-ная точка: {quantile_025:.3f} USD")
```

1. Математическое ожидание (среднее значение):

Средняя зарплата составляет 137,570.390 USD. Это значение указывает на то, что в среднем сотрудники получают около 137,570 USD в год.

2. Дисперсия:

Дисперсия зарплат равна 3,976,011,879.228 USD². Высокая дисперсия указывает на значительное рассеяние данных вокруг среднего значения, что говорит о большом разбросе зарплат.

3. Асимметрия:

Асимметрия равна 0.536. Положительная асимметрия указывает на то, что распределение зарплат скошено вправо, то есть имеет более длинный хвост справа. Это означает, что в выборке присутствуют несколько высокооплачиваемых сотрудников, которые увеличивают среднее значение.

4. Эксцесс:

Эксцесс равен 0.834. Положительный эксцесс указывает на то, что распределение зарплат имеет более острый пик по сравнению с нормальным распределением, но не настолько, чтобы считаться "толстохвостым".
